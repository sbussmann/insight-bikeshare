```python
import pandas as pd
import datetime
import numpy as np
```

```python
trips = pd.read_csv('hubway_trips_med.csv')
```

```python
trips.columns
```

```python
duration = trips[(trips['duration'] > 60) & (trips['duration'] < 3600)]['duration'].values
hist((duration/60.), bins=100)
```

```python
temp = pd.DatetimeIndex(trips['start_date'])
trips['start_day'] = temp.date
trips['start_time'] = temp.time
```

```python
trips['rideflag'] = np.ones(len(temp))
grouptrips = trips.groupby('start_day')
```

```python
tripsum = grouptrips.aggregate(np.sum)
```

```python
nriders = grouptrips.size()
```
