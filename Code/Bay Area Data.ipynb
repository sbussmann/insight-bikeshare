# Consolidate annual Bay Area data summaries into one csv file

```python
import pandas as pd
%matplotlib inline
```

```python
# read in the data
d1 = pd.read_csv('../Data/BayArea/201402_trip_data.csv')
d2 = pd.read_csv('../Data/BayArea/201408_trip_data.csv')
```

```python
dall = [d1, d2]
dconcat = pd.concat(dall)
```

```python
dconcat.to_csv('../Data/BayArea/alldata_bayarea.csv')
```
