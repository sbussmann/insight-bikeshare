# Consolidate quarterly Washington DC data summaries into one csv file

```python
import pandas as pd
%matplotlib inline
import glob
```

```python
# read in the data
datafiles = glob.glob('../Data/WashingtonDC/*History*csv')
datafiles
```

```python
dall = [pd.read_csv(datafile) for datafile in datafiles]
dconcat = pd.concat(dall)
```

```python
dconcat.to_csv('../Data/WashingtonDC/alldata_WashingtonDC.csv')
```
